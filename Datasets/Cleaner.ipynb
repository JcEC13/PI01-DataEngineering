```python
#Importando librerias necesarias
import pandas as pd
```

```python
#Lee todos los datasets disponibles
df_a=pd.read_csv('Dirtydata/amazon_prime_titles-score.csv')
df_d=pd.read_csv('Dirtydata/disney_plus_titles-score.csv')
df_h=pd.read_csv('Dirtydata/hulu_titles-score (2).csv')
df_n=pd.read_csv('Dirtydata/netflix_titles-score.csv')
```

Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

```python
df_a['show_id']='a'+df_a['show_id']
df_d['show_id']='d'+df_d['show_id']
df_h['show_id']='h'+df_h['show_id']
df_n['show_id']='n'+df_n['show_id']
```

Unir todos los DataFrames creados

```python
df=pd.concat([df_a,df_d,df_h,df_n], axis=0)
#Sentencia para comprobar que la union ha sido exitosa
print([len(df_a),len(df_d),len(df_h),len(df_n)])
print(len(df))
```

Los campos de texto deberán estar en minúsculas, sin excepciones

```python
#Creacion de una lista con con el nombre de las columnas que no son de tipo numericos para
#realizar su transformacion a minusculas
colums_str=df.columns.drop(['release_year','score'])
colums_str
```

```python
#Iterando la lista creada para transformar a minusculas cada una de las columnas
for i in colums_str:
    df[i]=df[i].str.lower()

df
```

Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”

```python
df.loc[df['rating'].isna(), 'rating'] = 'g'
df.head()
```

Las fechas deberán tener el formato AAAA-mm-dd

```python
df['date_added'] = pd.to_datetime(df.date_added)
df
```

El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
df[['duration_int', 'duration_type']] = df['duration'].str.split(' ', 1, expand=True)
df.loc[df['duration_type']=='seasons', 'duration_type'] = 'season'
df
```

```python
#Convierte a formato numerico a la columna duration_int para realizar consultas posteriores
df['duration_int'] = pd.to_numeric(df.duration_int, downcast='integer')
df.info()
```

```python
#Exportar el DataFrame con data procesada/limpia para su posterior consulta
df.to_csv('data.csv')
```
